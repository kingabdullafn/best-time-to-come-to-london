```python
mean_temp = np.mean(df['Temperature (C)'])
print("Mean temprature is ",mean_temp)
```

```python
var_temp = np.var(df['Temperature (C)'])
print("The variation of tempreature ", var_temp)
```

```python
standard_deviation_temp = np.std(df['Temperature (C)'])
print("Standard deviation of tempreature", standard_deviation_temp)
```

```python
for i in range(1,13):
  month= df.loc[df['month'] == i]['Temperature (C)']
  print("For month"+str(i))
  print("Mean tempreature is "+ str(np.mean(month)))
  print("Standard deviation is "+ str(np.std(month))+"\n")
```
